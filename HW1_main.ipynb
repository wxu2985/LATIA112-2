# Q1. 性別分布情形

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV文件
df = pd.read_csv('students_adaptability_level_online_education.csv')

# 計算Gender為Girl和Boy的人數
girl_count = len(df[df['Gender'] == 'Girl'])
boy_count = len(df[df['Gender'] == 'Boy'])

# 計算Gender總數
total_count = girl_count + boy_count

# 創建長條圖
labels = ['女生', '男生']
counts = [girl_count, boy_count]

plt.bar(labels, counts, color=['pink', 'skyblue'])

# 添加標題與軸標籤
plt.title('男女生人數統計')
plt.xlabel('性別')
plt.ylabel('人數')

# 顯示圖表
plt.show()


# 印出人數
print("\n分析結果:")
print("-------------------------------------------------------------------")
print("女生人數:", girl_count)
print("男生的人數:", boy_count)
print("總人數:", total_count)
print("-------------------------------------------------------------------")
```

#Q2. 男女比


```python
# 計算Boy和Girl的百分比
boy_percentage = (boy_count / total_count) * 100
girl_percentage = (girl_count / total_count) * 100


# 創建圓餅圖
labels = ['女生', '男生']
sizes = [girl_count, boy_count]
colors = ['pink', 'skyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140)

# 添加標題
plt.title('男女比')

# 顯示圖表
plt.axis('equal')
plt.show()


print("\n分析結果:")
print("-------------------------------------------------------------------")
print("男生的百分比: {:.2f}%".format(boy_percentage))
print("女生的百分比: {:.2f}%".format(girl_percentage))
print("-------------------------------------------------------------------")
```

#Q3. 經濟條件的分布情形

```python

# 計算每個Financial Condition類別的人數
financial_condition_counts = df['Financial Condition'].value_counts()

# 計算每個類別的百分比
poor_percentage = round((financial_condition_counts['Poor'] / total_count) * 100 ,2)
mid_percentage = round((financial_condition_counts['Mid'] / total_count) * 100 ,2)
rich_percentage = round((financial_condition_counts['Rich'] / total_count) * 100 ,2)

# 印出結果
print("分析數據:")
print("-------------------------------------------------------------------")
print("貧困人數:", financial_condition_counts['Poor'])
print("中等人數:", financial_condition_counts['Mid'])
print("富裕人數:", financial_condition_counts['Rich'])
print("總人數:", total_count)
print("-------------------------------------------------------------------")

# 創建圓餅圖
labels = ['貧困', '中等', '富裕']
sizes = [poor_percentage, mid_percentage, rich_percentage]
colors = ['pink', 'skyblue', 'gold']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140)

# 添加標題
plt.title('經濟狀況分佈')

# 顯示圖表
plt.axis('equal')
plt.show()

# 印出結果
print("\n分析結果:")
print("-------------------------------------------------------------------")
print("貧困百分比:", poor_percentage, "%")
print("中等百分比:", mid_percentage, "%")
print("富裕百分比:", rich_percentage, "%")
print("-------------------------------------------------------------------")
```

#Q4. 不同經濟條件(富有與貧窮)對線上教育適應度的影響

```python

# 使用條件篩選找出 Financial Condition 為 Rich 且 Adaptivity Level 為 High 的人數
rich_and_high_adaptivity = df[(df['Financial Condition'] == 'Rich') & (df['Adaptivity Level'] == 'High')]
num_rich_and_high_adaptivity = len(rich_and_high_adaptivity)


# 使用條件篩選找出 Financial Condition 為 Rich 且 Adaptivity Level 為 Low 的人數
rich_and_low_adaptivity = df[(df['Financial Condition'] == 'Rich') & (df['Adaptivity Level'] == 'Low')]
num_rich_and_low_adaptivity = len(rich_and_low_adaptivity)

# 使用條件篩選找出 Financial Condition 為 Poor 且 Adaptivity Level 為 High 的人數
poor_and_high_adaptivity = df[(df['Financial Condition'] == 'Poor') & (df['Adaptivity Level'] == 'High')]
num_poor_and_high_adaptivity = len(poor_and_high_adaptivity)

# 使用條件篩選找出 Financial Condition 為 Poor 且 Adaptivity Level 為 Low 的人數
poor_and_low_adaptivity = df[(df['Financial Condition'] == 'Poor') & (df['Adaptivity Level'] == 'Low')]
num_poor_and_low_adaptivity = len(poor_and_low_adaptivity)

# 計算 rich_and_low_adaptivity 與 rich_and_high_adaptivity 的倍數關係
ratio_rich = num_rich_and_high_adaptivity / num_rich_and_low_adaptivity

# 計算 poor_and_high_adaptivity 與 poor_and_low_adaptivity 的倍數關係
ratio_poor = num_poor_and_high_adaptivity / num_poor_and_low_adaptivity


# 輸出結果
print("分析數據:")
print("--------------------------------------------------------------------------")
print("在經濟條件為富裕的條件下: \n適應力高的人數:", num_rich_and_high_adaptivity,"\n適應力低的人數:", num_rich_and_low_adaptivity)
print("\n在經濟條件為貧困的條件下: \n適應力高的人數:", num_poor_and_high_adaptivity,"\n適應力低的人數:", num_poor_and_low_adaptivity)
print("\n註: 適應度中等的數據不列入分析")
print("--------------------------------------------------------------------------")

# 創建堆疊直條圖
labels = ['富裕', '貧困']
low_adaptivity_counts = [num_rich_and_low_adaptivity, num_poor_and_low_adaptivity]
high_adaptivity_counts = [num_rich_and_high_adaptivity, num_poor_and_high_adaptivity]

plt.bar(labels, low_adaptivity_counts, bottom=high_adaptivity_counts, color='orange', label='低適應度')
plt.bar(labels, high_adaptivity_counts, color='yellow', label='高適應度')

# 添加標題與軸標籤
plt.title('不同經濟條件對線上教育適應度的影響')
plt.xlabel('經濟條件')
plt.ylabel('人數')

# 添加圖例
plt.legend()

# 顯示圖表
plt.show()

print("\n分析結果:")
print("--------------------------------------------------------------------------")
print("經濟條件富裕的條件下，適應力「高」的人數是適應力「低」人數的 {:.2f} 倍".format(ratio_rich))
print("經濟條件貧困的條件下，適應力「高」的人數是適應力「低」人數的 {:.2f} 倍".format(ratio_poor))
print("\n所以可由上述的數據與圖表可得知富裕的學生對線上學習適應力較高，貧困的學生對線上學習適應力較低")
print("--------------------------------------------------------------------------")
```

# Q8. 住在城市與住在非城市所使用的網路型態差別

```python
# 使用條件篩選找出 Network Type 為 2G 的人數
network_2g = df[df['Network Type'] == '2G']
num_2g_yes = len(network_2g[network_2g['Location'] == 'Yes'])
num_2g_no = len(network_2g[network_2g['Location'] == 'No'])

# 使用條件篩選找出 Network Type 為 3G 的人數
network_3g = df[df['Network Type'] == '3G']
num_3g_yes = len(network_3g[network_3g['Location'] == 'Yes'])
num_3g_no = len(network_3g[network_3g['Location'] == 'No'])

# 使用條件篩選找出 Network Type 為 4G 的人數
network_4g = df[df['Network Type'] == '4G']
num_4g_yes = len(network_4g[network_4g['Location'] == 'Yes'])
num_4g_no = len(network_4g[network_4g['Location'] == 'No'])

# 印出結果
print("分析數據:")
print("-------------------------------------------------------------------")
print("住在城市使用2G人數有",num_2g_yes, "人, 住在非城市使用2G人數有",num_2g_no,"人")
print("住在城市使用3G人數有",num_3g_yes, "人, 住在非城市使用3G人數有",num_3g_no,"人")
print("住在城市使用4G人數有",num_4g_yes, "人, 住在非城市使用4G人數有",num_4g_no,"人")
print("-------------------------------------------------------------------")

# 創建堆疊長條圖
labels = ['2G', '3G', '4G']
counts_yes = [num_2g_yes, num_3g_yes, num_4g_yes]
counts_no = [num_2g_no, num_3g_no, num_4g_no]

plt.bar(labels, counts_no, bottom=counts_yes, color='deepskyblue', label='住在非城市')
plt.bar(labels, counts_yes, color='tomato', label='住在城市')


# 添加標題與軸標籤
plt.title('住在城市與住在非城市所使用的網路型態差別')
plt.xlabel('網路型態')
plt.ylabel('人數')

# 添加圖例
plt.legend()

# 顯示圖表
plt.show()

# 印出結果
print("\n分析結果:")
print("-------------------------------------------------------------------")
print("由圖表得知無論有沒有住在城市，都是使用4G的人數最多，使用2G的人數最少")
print("而在居住於非城市的族群中使用3G與4G的比例差異不大(藍色)")
print("但在居住於城市的族群中使用4G的比例明顯多於3G(紅色)")
print("-------------------------------------------------------------------")
```
